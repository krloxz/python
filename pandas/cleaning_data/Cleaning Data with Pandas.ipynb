## Dealing with categorical data

```python
# Creating data frame with categorical features only
cat_cols = [col for col in data_df.columns if data_df[col].dtype == "object"]
cat_df = data_df[cat_cols]
cat_df
```

```python
# Printing the unique values of each feature
for col in cat_df.columns:
  print(f"{col}:\n{cat_df[col].unique()}\n")
```

```python
# Manually encoding the labels of the arrival date month feature
cat_df.loc[:, "arrival_date_month"] = cat_df["arrival_date_month"].map({
  "January": 1,
  "February": 2,
  "March": 3,
  "April": 4,
  "May": 5,
  "June": 6,
  "July": 7,
  "August": 8,
  "September": 9,
  "October": 10,
  "November": 11,
  "December": 12
})
cat_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# Label-encode country and hotel
cat_df.loc[:, "country"] = encoder.fit_transform(cat_df["country"])
cat_df.loc[:, "hotel"] = encoder.fit_transform(cat_df["hotel"])

cat_df.head()
```

```python
# One-hot encoding
cat_df = pd.get_dummies(cat_df, dtype=int, columns=["meal", "market_segment", "distribution_channel",
                                                    "reserved_room_type",  "assigned_room_type", "deposit_type",
                                                    "customer_type", "reservation_status"])
cat_df
```

```python
num_df = data_df.drop(cat_cols, axis=1)
num_df
```

```python
final_df = pd.concat([num_df, cat_df], axis=1)
final_df
```

## Correlation Analysis

```python
corr_matrix = final_df.corr().abs()
corr_matrix
```

```python
import numpy as np

lower_triangle = corr_matrix.where(np.tril(np.ones(corr_matrix.shape), k=-1).astype(bool))
lower_triangle
```

```python
to_drop = []
for i in range(len(corr_matrix.columns)):
  for j in range(i):
    if abs(corr_matrix.iloc[i,j]) >= 0.85:
      to_drop.append(corr_matrix.columns[i])
to_drop
```

```python
# Drop columns with high correlation because they don't add information that could be useful for machine learning algorithms
# Having highly correlated data is like having this ecuation system:
# x + y = 1
# 2x + 2y = 2
# Both equations end up being the same and cannot be used to solve each other
final_df.drop(columns=to_drop, inplace=True)
final_df
```

```python
# We still need to clean up the agent column
final_df.agent.isnull().sum()
```

```python
# We fill the null values with 0z because the dataset creator used NULL to represent
# cases where the agent was not required, which makes it a legit category
# and 0 helps us to represent that category
final_df.agent = final_df.agent.fillna(0)
final_df.agent.isnull().sum()
```
