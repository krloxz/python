```python
houses = pd.read_csv('homes.csv')
houses.head()
```

```python
houses.plot()
```

```python
houses.plot(x='List', y='Rooms', marker='o', markerfacecolor='red')
```

```python
houses.plot(x='List', y='Rooms', kind='scatter')
```

```python
recent_houses = houses[houses['Age']<=50]
recent_houses.plot(kind='scatter', x='Age', y='List')
```

```python
# Remove the outlayer point
recent_houses[recent_houses['List']<=300].plot(kind='scatter', x='Age', y='List')
```

```python
cities = pd.read_csv('cities.csv')
cities_2019 = cities[cities['Year'] == 2019]
cities_2019
```

```python
cities_2019.plot(kind='bar', x='City', y='Population')
```

```python
cities_2019.plot(kind='barh', x='City', y='Population', title='2019 Population', legend=None)
```

```python
enrollment = pd.read_csv('enrollment.csv')
enrollment.head()
```

```python
enrollment.plot(kind='area', x='Year')
```

```python
pivot_enrollment = enrollment.pivot(index='Year', columns='Class', values='Enrollment')
pivot_enrollment
```

```python
pivot_enrollment.plot(kind='area')
```

```python
pivot_enrollment.plot(kind='area', stacked=False)
```
