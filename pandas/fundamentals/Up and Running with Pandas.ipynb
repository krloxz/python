## Basics

```python
import pandas as pd
```

```python
df_sales = pd.DataFrame(columns=['Date', 'Amount'])
df_sales
```

```python
df_new_data = pd.DataFrame({
  'Date': ['January 1', 'January 2'],
  'Amount': [500.00, 600.00]
})
df_new_data
```

```python
df_sales = pd.concat([df_sales, df_new_data])
df_sales
```

```python
df_second = pd.DataFrame({
  'Date': ['January 3', 'January 4'],
  'Amount': [100.00, 200.00]
})
df_second
```

```python
df_new = pd.concat([df_sales, df_second])
df_new
```

```python
df_new.reset_index(drop=True, inplace=True)
df_new
```

```python
df_new_column = pd.DataFrame({
  'Location': ['Store A', 'Store B']
})
df_new = pd.concat([df_new, df_new_column], axis=1)
df_new
```

```python
df_headers = pd.DataFrame(columns=['Date', 'Branch', 'Amount', 'Active'])
df_data = pd.DataFrame({
  'Date': ['January 1', 'January 2', 'January 3'],
  'Branch': ['Branch A', 'Branch B', 'Branch C'],
  'Amount': [100.00, 200.00, 300.00],
  'Active': [True, False, True]
})
pd.concat([df_headers, df_data])
```
