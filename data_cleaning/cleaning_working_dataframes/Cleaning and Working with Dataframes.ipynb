## Reshaping and Pivoting Dataframes

```python
df = pd.DataFrame({
  'purchase_date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03'],
  'product': ['A', 'B', 'A', 'B', 'A', 'B'],
  'price': [1, 2, 3, 4, 5, 6]
})
df
```

```python
df_pivoted = df.pivot(columns='product', index='purchase_date', values='price')
df_pivoted
```

```python
df_pivoted.reset_index().melt(id_vars='purchase_date', value_vars=['A', 'B'])
```

## Merging and Joining Dataframes

```python
df1 = pd.DataFrame({
  'course': ['A', 'B', 'C', 'D'],
  'teachers': [1, 2, 3, 4]
})
df1
```

```python
df2 = pd.DataFrame({
  'course': ['C', 'D', 'E', 'F'],
  'students': [5, 6, 7, 8]
})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='course')
```

```python
pd.merge(df1, df2, on='course', how='inner')
```

```python
pd.merge(df1, df2, on='course', how='left')
```

```python
pd.merge(df1, df2, on='course', how='right')
```

```python
pd.merge(df1, df2, on='course', how='outer')
```

```python
pd.merge(df1, df2, how='cross')
```

## Altering Data Types

```python
df = pd.DataFrame({
  'Order_Id': [102, 103, 104],
  'Order_Date': ['01/10/2023', '05/10/2023', '10/10/2023'],
  'Product_Id': ['001', '002', '003'],
  'Price': ['10.5', '5.3', '7.8'],
  'Units_Sold': ['100', '150', '120'],
  'Color': ['Red', 'Blue', 'Green'],
  'Name': ['John', 'Emily', 'Michael'],
  'Age': ['25', '30', '35'],
  'City': ['New York', 'Paris', 'London'],
})
df
```

```python
df.dtypes
```

```python
df['Age'] = df['Age'].astype(int)
df['Price'] = df['Price'].astype(float)
df['Units_Sold'] = df['Units_Sold'].astype(int)
df.dtypes
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df.dtypes
```

```python
df['Color'] = df['Color'].astype('category')
df.dtypes
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d/%m/%Y')
df.dtypes
```

## Replacing Values

```python
df = pd.DataFrame({
  'Product': ['A', 'B', 'C', 'D', 'E'],
  'Rating': ['Good', 'Average', 'Poor', 'Average', 'Excellent']
})
df
```

```python
df['Rating'] = df['Rating'].replace('Excellent', 'Good')
df
```

```python
df['Rating'] = df['Rating'].replace({'Good': 'Positive', 'Average': 'Neutral', 'Poor': 'Negative'})
df
```

## Data Aggregation

```python
df = pd.DataFrame({
  'Region': ['East', 'West', 'East', 'North', 'West', 'North'],
  'Product': ['A', 'A', 'B', 'A', 'B', 'B'],
  'Sales': [100, 150, 200, 50, 300, 250]
})
df
```

```python
df.groupby('Region')['Sales'].sum()
```

```python
df.groupby('Region')['Sales'].aggregate(['sum', 'mean', 'max', 'min'])
```

```python
df.groupby(['Region', 'Product'])['Sales'].sum()
```

```python
df.groupby(['Region', 'Product'])['Sales'].sum().unstack()
```
